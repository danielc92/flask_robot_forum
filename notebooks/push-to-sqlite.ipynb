# Imports 

```python
import pandas 
import sqlite3
```

# Create connection to db

```python
uri = 'robot.db'
engine = sqlite3.connect(uri)
```

# Import all robot data

```python
robots = pandas.read_csv('./forum-data/robots.csv', sep = '|')
threads = pandas.read_csv('./forum-data/threads.csv', sep = '|')
comments = pandas.read_csv('./forum-data/comments.csv', sep = '|')
```

# Preview

```python
robots.head()
```

```python
threads.head()
```

```python
comments.head()
```

# Push into database

```python
robots.to_sql('robots', con = engine, index = False, if_exists='replace')
```

```python
threads.to_sql('threads', con = engine, index = False, if_exists='replace')
```

```python
comments.to_sql('comments', con = engine, index = False, if_exists='replace')
```
